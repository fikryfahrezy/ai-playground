```python
import os
import boto3 
from dotenv import load_dotenv

load_dotenv()

bedrock = boto3.client(service_name="bedrock")
bedrock_runtime = boto3.client(service_name="bedrock-runtime")

for model in bedrock.list_foundation_models(
    byCustomizationType="FINE_TUNING")["modelSummaries"]:
    for key, value in model.items():
        print(key, ":", value)
    print("-----\n")
```

```python
job_name = "my-fine-tunning-model-job-3"

# Select the foundation model you want to customize
base_model_id = "cohere.command-light-text-v14:7:4k"

bedrock.create_model_customization_job(
    customizationType="FINE_TUNING",
    jobName=job_name,
    customModelName="my-fine-tunning-model",
    roleArn=os.getenv("ROLE_ARN"),
    baseModelIdentifier=base_model_id,
    hyperParameters = {
        "epochCount": "1",
        "batchSize": "8",
        "learningRate": "0.00001",
    },
    trainingDataConfig={"s3Uri": "s3://my-fine-tunning-bucket/train-summarization.jsonl"},
    outputDataConfig={"s3Uri": "s3://my-fine-tunning-bucket/trainned-summarization"},
)

# Check for the job status
status = bedrock.get_model_customization_job(jobIdentifier=job_name)["status"]
print(status)
```
